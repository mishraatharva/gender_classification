```python
data = pd.read_csv(r'U:\nlp_project\gender_classification\data\cleaned_data.csv')
```

```python
data['new_data'] = data['description'].str.cat(data['text'], sep=' ')
```

```python
cv = CountVectorizer(max_features = 1500)
x = cv.fit_transform(datanew['new_data']).toarray()
```

```python
x
```

```python
encoder = LabelEncoder()
y = encoder.fit_transform(datanew['gender'])
```

```python
pd.DataFrame(y).value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

### MultinomialNB

```python
nb = MultinomialNB()
nb.fit(x_train, y_train)
```

```python
print(nb.score(x_test, y_test))
```

```python
y_pred = nb.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
confusion_matrix(y_test, y_pred)
```

### LogisticRegression

```python
lr = LogisticRegression(max_iter = 2000)
lr.fit(x_train,y_train)
# prediction
y_pred = lr.predict(x_test)
accuracy = 100.0 * accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```

### RandomForestClassifier

```python
rf_classifier = RandomForestClassifier(n_estimators=100,random_state=42)
rf_classifier.fit(x_train,y_train)
```

```python
y_pred = rf_classifier.predict(x_test)
accuracy = 100.0 * accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```

### GradientBoostingClassifier

```python
gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=0.05,random_state=42)
```

```python
gbc.fit(x_train,y_train)
```

```python
y_pred = gbc.predict(x_test)
accuracy = 100.0 * accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```

### AdaBoostClassifier

```python
abc = AdaBoostClassifier(n_estimators=100, random_state=42,)
abc.fit(x_train,y_train)
```

```python
y_pred = abc.predict(x_test)
accuracy = 100.0 * accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```
